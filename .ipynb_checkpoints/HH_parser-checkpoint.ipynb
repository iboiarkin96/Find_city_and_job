```python
# функция которая парсит первую страницу в опредленном городе и по опредленной вакансии
def getPage(page = 0, vacancy = 'NAME:Аналитик', area = 1):
    """
  
    page - Индекс страницы, начинается с 0. Значение по умолчанию 0, т.е. первая страница
    vacancy = 'NAME:Аналитик',  - вакансия аналитик, это Текст фильтра
    area = 1 - Поиск ощуществляется по вакансиям города Москва
    """
     
    # Справочник для параметров GET-запроса
    params = {
        'text': vacancy, 
        'area': area, 
        'page': page, # 
        'per_page': 100 # Кол-во вакансий на 1 странице
    }
     
     
    req = requests.get('https://api.hh.ru/vacancies', params) # Посылаем запрос к API
    data = req.content.decode() # Декодируем его ответ, чтобы Кириллица отображалась корректно
    req.close()
    return data
 
#  функция созадния директории    
def create_folder(workspace, folder):
    path = os.path.join(workspace, folder)
    if not os.path.exists(path):
        return os.makedirs(path) 
    
# функция которая создает директорию, просматривает указаное колличесво страниц на сайте hh
# для одного челоека на одну вакансию  в одном городе  
def getting_vacation(names = 'Иван', count_of_page = 200, 
                     page = 0, vacancy = 'NAME:Аналитик', area = 1):
    
    '''
    names = ['Иван'] - человек по которому ведется поиск
    count_of_page = 2 - максимальная глубина поиска
    другие аргументы повторяются 
    '''


#     for name_family_member in tqdm(names) :
#     проверяем дирикторию, если такой директории нет, то создаем папку
    if  os.path.exists(os.getcwd() + '/Вакансии/' + names):
        pass
    else:
        create_folder(os.getcwd() + '/Вакансии/', names)
    for page in range(0, count_of_page):
        try:

            # Преобразуем текст ответа запроса в справочник Python
            jsObj = json.loads(getPage(page, vacancy, area))
            if int(jsObj['found'])==0 :
                    break
#           ограничиваем добавление тустых jsons
            elif len(jsObj['items']) !=0:
        
                # Сохраняем файлы в папку {путь до текущего документа со скриптом}\docs\pagination
                nextFileName = str(os.getcwd()) \
                    + '/Вакансии/{}/'.format(names) \
                    + '{}.json'.format(names 
                                        + "_" \
                                        +vacancy.split(':')[1:][0] 
                                        + "_city " \
                                        +str(area) 
                                       + "_page " \
                                        + str(page + 1) )

                # Создаем новый документ, записываем в него ответ запроса, после закрываем
                f = open(nextFileName, mode='w', encoding='utf8')
                f.write(json.dumps(jsObj, ensure_ascii=False))
                f.close()
                # Проверка на последнюю страницу, если вакансий меньше 2000
                if (jsObj['pages'] - page) <= 1:
                    break
                
        except:
            break

        # Необязательная задержка, но чтобы не нагружать сервисы hh, оставим. 5 сек мы может подождать
        time.sleep(0.25)
```

```python
all_territoruy = 'https://api.hh.ru/areas'
req = requests.get(all_territoruy) # Посылаем запрос к API
data = req.content.decode()

def extract_code_of_terr():
    
    type_1 = "eval(data.replace('false', 'False').replace('true', 'True').replace('null', 'None'))"
    type_2 = type_1 + "[0]['areas']"
    
    box_type_1 = []
    box_type_2 = []
    
    for type_, box_ in zip([type_1, type_2], [box_type_1, box_type_2]):
        full_dict = eval(type_)
        
        try:
            for j in range(10000):
                try:
                    for i in range(10000):
                        box_.append(
                                    [
                                    full_dict[j]['name'], 
                                    full_dict[j]['areas'][i]['id'], 
                                    full_dict[j]['areas'][i]['name']
                                    ]
                        )
                except:
                    ...
        except:
            ...
    df1= pd.DataFrame(box_type_1, columns =['name_of_regin', 'code_of_city', 'name_of_city']).query('name_of_regin == "Россия"')
    df2= pd.DataFrame(box_type_2, columns =['name_of_regin', 'code_of_city', 'name_of_city'])
    
    df_new =  pd.merge(df1, df2,  left_on=['name_of_city'], right_on=['name_of_regin'], how = 'left')
    df_new.drop(columns =['name_of_regin_y'] , inplace = True)
    df_new.rename(columns = {'name_of_regin_x':'Страна', 
                            'code_of_city_x':'Код_региона', 
                            'name_of_city_x':'Название_региона', 
                            'code_of_city_y':'Код_города', 
                            'name_of_city_y':'Название_города' },  inplace=True)
    
    for row in range(df_new.shape[0]):
        if df_new.loc[row, 'Код_города'] != df_new.loc[row, 'Код_города']:
            df_new.loc[row, 'Код_города'] = df_new.loc[row, 'Код_региона']
            df_new.loc[row, 'Название_города'] = df_new.loc[row, 'Название_региона']
    
    df_new['Код_города'] = df_new['Код_города'].astype('int64')
    df_new['Код_региона'] = df_new['Код_региона'].astype('int64')
    
    return df_new.sort_values(by = 'Код_региона').reset_index(drop = True)

all_city = extract_code_of_terr()
```

Сходство документов вычисляется путем вычисления расстояния между документами. Расстояние до документа-это понятие, в котором слова(документы) рассматриваются как векторы и вычисляются как угол между двумя заданными векторами документа. Векторы документа - это частота встречаемости слов в данном документе. Давайте рассмотрим пример:

```python
stop_words = get_stop_words('russian')
nltk.download('punkt') # if necessary...
```

```python
stemmer = nltk.stem.porter.PorterStemmer()
remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)

def stem_tokens(tokens):
    return [stemmer.stem(item) for item in tokens]

'''remove punctuation, lowercase, stem'''
def normalize(text):
    return stem_tokens(nltk.word_tokenize(text.lower().translate(remove_punctuation_map)))



def cosine_sim(text1, text2, name = "Иван"):
    if name == "Иван":
        stop_= stop_words + stop_words_ivan
    else:
        stop_ = stop_words + stop_words_natasha
    vectorizer = TfidfVectorizer(tokenizer=normalize, stop_words=stop_)
    tfidf = vectorizer.fit_transform([text1, text2])
    return ((tfidf * tfidf.T).A)[0,1]
        
```

# Прикручиваем сходство текста

```python
stop_words_ivan, stop_words_natasha = [['Нефть', "R", 'Kafka'],
                                                                           
                                        [ "Английский язык"]]
```

```python
expectations_Ivan = ''' SQL. Базовые знания реляционных СУБД (MSSSQL, ClickHouse).
                        Высшее образование в области анализа данных, математики, статистики
                        Проведение A/B тестов
                        Знание основных Python библиотек для ML, анализа данных и визуализации
                        Уверенное владение Python (numpy, scipy, pytorch, sklearn, matplotlib)
                        машинное обучение'''


expectations_Natasha = '''Высшее образование в области химии, управление химихо-технологическими предприятия.
                        физико-химические методы анализа.
                         Cборки лабораторных установок.
                        Использование контрольно-измерительными приборами, аппаратов.
                        Анализ нефтепродуктов'''
```

```python
def measurement(df_final, expectations_Ivan, expectations_Natasha):
    box = []
    df_final['score_similarity'] = 0
    for i in tqdm(range(df_final.shape[0])):
        name = df_final.loc[i,'name_family']
        description= df_final.loc[i,'key_skills']
        if name == "Иван":
            score = cosine_sim(description, expectations_Ivan, name = name)
        else:
            score = cosine_sim(description, expectations_Natasha, name = name)
        box.append(score)
    return box
```

```python
score = measurement(df_final, expectations_Ivan, expectations_Natasha)
```

```python
df_final['score_similarity']= score
df_final = df_final.query('score_similarity > 0')
```

```python
f, axes = plt.subplots(1, 1, figsize = (15, 5))
sns.histplot(data =df_final,
             x = 'score_similarity',
             hue = 'name_family')
```

```python
def plot_by_regin(df_final):
    for region in df_final['Название_региона'].unique():
        data = df_final[df_final['Название_региона']==region]
        if data.shape[0] > 3:
            f, axes = plt.subplots(1, 2, figsize = (25, 5))
            for ind, wage in enumerate(['salary.from', 'salary.to']):
                plt.suptitle(region)
                sns.histplot(data = data,
                            x = wage,
                            hue = 'name_family',
                            ax =axes[ind],
                            multiple="dodge", shrink=.9)
                axes[ind].set_title(f"Распределение {wage}")
```
